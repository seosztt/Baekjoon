```python
d=list('ABCDEFGHIJKLMNOPQRTUVWXY')
n=list('23456789')
dic={}
dic.update(list(zip(d[::3],n)))
dic.update(list(zip(d[1::3],n)))
dic.update(list(zip(d[2::3],n)))
dic.update([('S','7'),('Z','9')])
a=input()
time=0
for i in range(len(a)):
    time+=(int(dic[a[i]])+1)
print(time)
```
