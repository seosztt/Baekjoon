```python
def star(n):
    if n==1:
        return ['*']
    gen=star(n//3)
    l=[]
    for s in gen:
        l.append(s*3)
    for s in gen:
        l.append(s+' '*(n//3)+s)
    for s in gen:
        l.append(s*3)
    return l
n=int(input())
print('\n'.join(star(n)))
```
