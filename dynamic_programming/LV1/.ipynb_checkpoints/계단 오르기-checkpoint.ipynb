```python
n=int(input())
L=[0]
for _ in range(n):
    L.append(int(input()))
    
d=[0]*(n+1)

if n==1:
    print(L[1])
elif n==2:
    print(L[1]+L[2])
else:
    d[1]=L[1]
    d[2]=L[1]+L[2]
    d[3]=L[3]+max(L[1], L[2])

    for i in range(4,n+1):
        d[i]=max(L[i]+L[i-1]+d[i-3], L[i]+d[i-2])
    print(d[n])
```
